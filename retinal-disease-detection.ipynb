```python
from ultralytics import YOLO
from sklearn.metrics import classification_report, confusion_matrix
import os
from PIL import Image
import torch
from torchvision import datasets, transforms
import numpy as np

# Load trained model
model = YOLO('/kaggle/working/runs/classify/train/weights/best.pt')  # or your best model path

# Set path to test or validation set
test_dir = '/kaggle/input/dataset/test'  # or 'dataset/valid'

# Define transforms (match your training ones)
transform = transforms.Compose([
    transforms.Resize((640, 640)),
    transforms.ToTensor(),
])

# Load dataset
test_dataset = datasets.ImageFolder(test_dir, transform=transform)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=32, shuffle=False)

# Class mapping
idx_to_class = {v: k for k, v in test_dataset.class_to_idx.items()}

# Run prediction
y_true, y_pred = [], []
for imgs, labels in test_loader:
    results = model(imgs, verbose=False)
    preds = torch.stack([
        r.probs.data.clone().detach().cpu().argmax() for r in results
    ])
    y_true.extend(labels.cpu().numpy())
    y_pred.extend(preds.cpu().numpy())

# Evaluation
print("Classification Report:")
print(classification_report(y_true, y_pred, target_names=list(idx_to_class.values())))

print("Confusion Matrix:")
print(confusion_matrix(y_true, y_pred))
```
